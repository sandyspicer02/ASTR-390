```python
import numpy as np
import matplotlib.pylab as plt
from scipy.optimize import curve_fit

%matplotlib inline
```

```python
def quadfunc(x1,m,b):
    return (m*x1)+b
period = np.array([1.769,3.551,7.155,16.689])
distance = np.array([421.7e3,670.9e3,1070.4e3,1882.7e3])
y1err = 0.001
x2 = np.log10(period)
y2 = np.log10(distance)
plt.figure()
plt.errorbar(x2,y2,yerr=y1err,fmt='o',label='First Set of Points')
fitresults,fituncertainties = curve_fit(quadfunc,x2,y2,sigma=y1err)
xfitpts = np.linspace(0,12)
fitpolynomial = plt.poly1d(fitresults)
yfitpts = fitpolynomial(xfitpts)
plt.plot(xfitpts,yfitpts,'r-',label='Fit results')
# plt.xlim(0,2)
# plt.ylim(2.5,3.5)
plt.xlabel('Log Period (d)')
plt.ylabel('Log Distance (m)')
plt.legend(loc='upper left')
plt.title('Jupiters Moons')
m = (y2[1]-y2[0])/(x2[1]-x2[0])
print m
```

```python
period = np.array([152841.6,306806.4,618192,1441929.6])
distance = np.array([421.7e6,670.9e6,1070.4e6,1882.7e6])

log_period = np.log10(period)
log_distance = np.log10(distance)

plt.plot(log_distance,log_period,'ro')
plt.xlabel('log(a)(m)')
plt.ylabel('log(P)(s)')
plt.xlim()
plt.title('Log (period) vs. Log (distance) for 4 Moons of Jupiter')
c1=np.polyfit(log_distance,log_period,1)
f1=np.linspace(min(log_distance),max(log_distance),100)
ffit=np.polyval(c1,f1) 
plt.plot(f1,ffit,'r')
c1
plt.savefig('logvlog.png')
```
