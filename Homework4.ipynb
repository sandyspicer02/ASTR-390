Your assignment is to generate one plot which shows the two stellar spectra (with suitably labeled axes) and a blackbody curve for each star whose temperature best matches (by eye, or quantitatively if you're so inclined!) the shape of each stellar spectrum.
```python
from __future__ import division 
import math
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
wavelength1, flux1 = np.loadtxt('star1.txt',unpack=True)
wavelength2, flux2 = np.loadtxt('star2.txt',unpack=True)

print(max(flux1)) # Corresponds to 4505 Angstroms
print(max(flux2)) # Corresponds to 2595 Angstroms

# Wien's Law

T1 = 2.90e-3/(4505*1e-10)
T2 = 2.90e-3/(2595*1e-10)

# F1 = flux1*1e14
# F2 = flux2*1e14

print T1
print T2
```

```python
def blackbody(wavelength,T):
    h = 6.626e-34
    c = 2.9979e8
    k = 1.3806e-23
    Lambda = wavelength*1e-10
    b = ((2*h*c**2)/(Lambda**5))*((1)/(math.e**((h*c)/(Lambda*k*T))-1))
    return b
```

```python
BB1 = []
BB2 = []

BB1 = blackbody(wavelength1,T1)
BB2 = blackbody(wavelength2,T2)
maxB1 = max(BB1)
maxB2 = max(BB2)

print maxB1
print maxB2
```

```python
plt.plot(wavelength1,flux1,'r-',label='Star 1')
plt.plot(wavelength2,flux2,'b-',label='Star 2')
plt.plot(wavelength1,BB1/maxB1,'g',label='Blackbody Spectrum for Star 1')
plt.plot(wavelength2,3.5*BB2/maxB2,'k',label='Blackbody Spectrum for Star 2')
plt.xlabel('Wavelength (Angstroms)')
plt.ylabel('Intensity')
plt.title('Stellar Spectra')
plt.legend(loc='upper right')
plt.savefig('stellar_spectra.png')
```
