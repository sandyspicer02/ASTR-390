Reproduce Figure 15.3 in the textbook (binding energy per nucleon vs atomic mass) using the attached file binding.txt.

```python
from __future__ import division 
import math
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
E,A = np.loadtxt('binding.txt',skiprows=2,usecols=(2,3),unpack=True)
```

```python
plt.plot(np.log(A),E)
plt.title('Binding Energy per Nucleon vs. Mass Number')
plt.xlabel('A')
plt.ylabel('E/A (MeV)')
plt.savefig('E_vs_A.png')
```
