```python
import numpy as np
import matplotlib.pylab as plt
import math
%matplotlib inline
```

```python
def velocity(e):
    l=1
    a=1
    phi = np.linspace(0,2*np.pi, 400)
    vx = -l*(np.sin(phi))/(a*(1-e**2))
    return vx

def orbital_period(e):
    phi = np.linspace(0,2*np.pi, 400) 
    t = 1/(2*np.pi)*(2*np.arctan(np.sqrt((1-e)/(1+e))*np.tan(phi/2))-((e*np.sqrt(1-e**2)*np.sin(phi))/(1+e*np.cos(phi))))+0.5
    return t
```

```python
# When the eccentricity is 0

vx = velocity(0)
#print vx

t = orbital_period(0)
#print t
```

```python
plt.plot(t,vx,'ro')
plt.xlabel('Time t/P')
plt.ylabel('Velocity in x-direction  (m/s)')
plt.title('Velocity Curve for e=0')
plt.savefig('e=0.png')
```
