```python
'''
First part is to generate a log-log plot of radius versus mass, both relative to the solar
values. Fit a power-law relationship to the so-called mass-radius relation and compare
your results to the “accepted” value.
'''
#NOW TAKE LOG10 OF ALL THE M FINALS


logm0_5final = np.log10(m0_5final)
logm1final = np.log10(m1final)
logm3final = np.log10(m3final)
logm8final = np.log10(m8final)
logm15final = np.log10(m15final)



#only use the last data point
```

```python
plt.figure()
plt.plot(logm0_5final, logradius0_5[first05[0]],'bo',label='$mass=0.5M\odot$') 
plt.plot(logm1final, logradius1[first1[0]], 'ro',label='$mass=1M\odot$')
plt.plot(logm3final, logradius3[first3[0]], 'mo',label='$mass=3M\odot$')
plt.plot(logm8final, logradius8[first8[0]], 'go',label='$mass=8M\odot$')
plt.plot(logm15final, logradius15[first15[0]], 'ko',label='$mass=15M\odot$')

#MAKE MODEL POWER

x = np.array([logm0_5final, logm1final,logm3final, logm8final, logm15final])
y = np.array([logradius0_5[first05[0]],logradius1[first1[0]],logradius3[first3[0]],logradius8[first8[0]],logradius15[first15[0]]])
                                    
bestfit = np.polyfit(x, y, 1)
print bestfit

#x= x
#y = mx+b where m is the output of polyfit and y-int is the second output\
#[0.83948494  1.86122528]

plt.plot(x,bestfit[0]*x+bestfit[1],label='$\\alpha =0.67$')

plt.legend(loc='best')
#plt.axis([-3,2,3.5,4.6])
plt.xlabel('$log(mass)$')
plt.ylabel('$log(radius)$')
plt.title('$log-log\ of\ radius\ vs.\ mass$')
plt.savefig('RvM.jpg')


#INCLUDE THE 3RD ONE NOW AND THE SLOPE AND Y-INT IS GONNA CHANGE AS WELL. 

#Accepted 
#R=M^alpha where my alpha here is my m or slope that i got from polyfit
```
